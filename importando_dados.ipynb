## Exercício: Quantas candidaturas houveram no seu município nas eleições de 2022?

```python
candidaturas_se = pd.read_csv('/content/consulta_cand_2022_SE.csv', sep=';', encoding='latin_1')
pd.options.display.max_columns = None
# candidaturas_se.head()

###  Quantos candidatos são naturais de Aracaju?
###   - Variável chave 'NM_MUNICIPIO_NASCIMENTO'
###   - Usar value.counts() para contar valores de uma coluna.
frequencia_cidades = candidaturas_se['NM_MUNICIPIO_NASCIMENTO'].value_counts()
```

```python
import seaborn as sns
sns.barplot(x=frequencia_cidades.index, y=frequencia_cidades.values)
```

```python
import matplotlib.pyplot as plt

top_5_cidades = frequencia_cidades.head(5)
sns.barplot(x=top_5_cidades.index, y=top_5_cidades.values)
plt.title('Top 5 Cidades com Mais Candidatos')
plt.xlabel('Cidade')
plt.ylabel('Número de Candidatos')
plt.xticks(rotation=45) # Rotaciona os rótulos do eixo x, se necessário
plt.show()
```

## Aula 7: Carregando tabelas

É possível ler vários arquivos CSV e converter de uma só vez para dataframe Pandas. Para isso, além de usar as técnicas vistas nas aulas anteriores, importaremos a biblioteca glob.

A biblioteca glob no Python é usada para encontrar todos os caminhos que correspondem a um padrão especificado, de acordo com as regras usadas pelo shell Unix. Isso é útil quando você precisa trabalhar com vários arquivos que têm nomes semelhantes ou padrões em seus nomes.

A função mais comumente usada no módulo glob é glob.glob(). Aqui está um exemplo de como ela pode ser usada:

```python
import glob
```

```python
#caminho_do_arquivo = '/content/consulta_cand_2022_BRASIL.csv'
#dados_todos_estados = pd.read_csv(caminho_do_arquivo, sep=';', encoding='latin_1')
# print(dados_todos_estados)


dados_todos_estados = glob.glob('conjunto_dados/consulta_*.csv')
print(dados_todos_estados)
```

Vamos ler todos os arquivos e reunir em uma lista.

```python
lista_df = []

for dados_estado in dados_todos_estados:
    df_estado = pd.read_csv(dados_estado, sep=';', encoding='latin_1')
    lista_df.append(df_estado)
```

Com o método `pd.concat` poderemos reunir os dataframes de cada estado em um único

```python
df_completo = pd.concat(lista_df, axis=0, ignore_index=True)
df_completo.shape
```
